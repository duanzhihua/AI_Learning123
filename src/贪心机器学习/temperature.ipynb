##  线性回归(作业)

### Part 1 导入相关包

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Part 2 读取数据

```python
data = pd.read_csv("../../data/height.vs.temperature.csv")
```

```python
data.head()
```

```python
data.columns
```

### Part 3 数据可视化

```python
plt.figure(figsize=(16,8))
plt.scatter(data["height"], data["temperature"], c='g')
plt.xlabel("Height")
plt.ylabel("Temperature")
plt.show()
```

### Part 4： 训练线性回归模型

```python
X= data["height"].values.reshape(-1,1)
y= data["temperature"].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X,y)
```

```python
print('w = {:.5}'.format(reg.coef_[0][0]))
print('b = {:.5}'.format(reg.intercept_[0]))
print('线性模型为：Y = {:.5}*X + {:.5}'.format(reg.coef_[0][0], reg.intercept_[0]))
```

### Part 5：模型可视化

```python
predictions = reg.predict(X)
plt.figure(figsize=(16,8))
plt.scatter(data["height"], data["temperature"], c='g')
plt.plot(data["height"], predictions,c='red',linewidth=2)
plt.xlabel("height")
plt.ylabel("temperature")
plt.show()
```

### Part 6：做预测
假设山的高度为8000米，那么预测温度是多少？

```python
predictions = reg.predict([[8000]])
print('山的高度为8000米, 预测温度是{:.5}度'.format(predictions[0][0]))
```
